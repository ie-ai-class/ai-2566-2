```python
from matplotlib import pyplot as plt
from aeon.datasets import load_longley
from aeon.forecasting.var import VAR
from aeon.utils.plotting import plot_series
```

```python
_, y = load_longley()
```

```python
y.head()
```

```python
y = y.drop(columns=['UNEMP', 'ARMED', "POP"])
```

```python
y
```

```python
plot_series(y['GNPDEFL'])
```

```python
plot_series(y['GNP'])
```

```python
forecaster = VAR()
```

```python
forecaster.fit(y)
```

```python
y_pred = forecaster.predict(fh = [1,2,3, 4])
```

```python
plot_series(y['GNP'], y_pred['GNP'])
```

```python
plot_series(y['GNPDEFL'], y_pred['GNPDEFL'])
```
